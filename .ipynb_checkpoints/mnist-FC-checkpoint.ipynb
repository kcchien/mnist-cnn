```python
import numpy as np
from keras.utils import np_utils
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape)
x_train_reshape = x_train.reshape(60000, 784).astype('float32')
print(x_train_reshape.shape)
x_train_normalized = x_train_reshape / 255
print(x_train_normalized[:5])
```

```python
y_train_onehot = np_utils.to_categorical(y_train)
print(y_train_onehot[:5])
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=256, input_dim=784, kernel_initializer="normal", activation="relu"))
```

```python
model.add(Dense(units=10, kernel_initializer="normal", activation="softmax"))
```

```python
print(model.summary())
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
train_history = model.fit(x=x_train_normalized, y=y_train_onehot, validation_split=0.2, epochs=10, batch_size=200, verbose=2)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
def show_train_history(train_history, train, validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title("Train History")
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.show()
```

```python
show_train_history(train_history, "acc", "val_acc")
```

```python
show_train_history(train_history, "loss", "val_loss")
```

```python
print(len(y_test))
```

```python
x_test_reshape = x_test.reshape(10000, 784).astype("float32")
x_test_normalized = x_test_reshape / 255
y_test_onehot = np_utils.to_categorical(y_test)
```

```python
scores = model.evaluate(x_test_normalized, y_test_onehot)
```

```python
print("Accuracy: {}%".format(scores[1]))
```

```python
import itertools
def plot_confusion_matrix(cm, classes, normalize=False, title="Confusion Matrix", cmap=plt.cm.Blues):
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    thresh = cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
results = model.predict_classes(x_test_reshape)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, results)
plot_confusion_matrix(cm, range(0, 9))
```

```python
incorrect = np.where(y_test != results)[0]
for i in range(0, 9):
    plt.subplot(3, 3, i+1)
    idx = incorrect[i]
    img = x_test[idx]
    plt.imshow(img, cmap="gray")
    plt.title("{}/{}".format(y_test[idx], results[idx]))
```

```python
correct = np.where(y_test == results)[0]
for i in range(0, 9):
    plt.subplot(3, 3, i+1)
    idx = correct[i]
    img = x_test[idx]
    plt.imshow(img, cmap="gray")
    plt.title("{}".format(y_test[idx]))
```
